```python
import requests
import matplotlib.pyplot as plt

def fetch_weather_data(api_url):
    try:
        response = requests.get(api_url)

        # Check if the request was successful
        if response.status_code == 200:
            data = response.json()  # Parse JSON data
            return data
        else:
            print(f"Error: Unable to fetch data. Status code {response.status_code}")
            return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# API URL
api_url = "http://api.worldweatheronline.com/premium/v1/weather.ashx?key=007f8f538a464794a8f115446250803&q=Cairo&format=json&num_of_days=7"

# Fetch the weather data
weather_data = fetch_weather_data(api_url)

# If data is successfully fetched, print it
if weather_data:
    # Extract and print the location of the city
    city_location = weather_data.get('data', {}).get('request', [{}])[0].get('query', 'Location not found')
    print(f"Location of the city: {city_location}")

    # Lists to store dates and chance of rain
    dates = []
    chance_of_rain_list = []

    # Extract and print the weather details for each day
    for weather in weather_data.get('data', {}).get('weather', []):
        date = weather.get('date', 'Date not found')
        dates.append(date)

        daily_chance_of_rain = 0
        for period in weather.get('hourly', []):
            chance_of_rain = period.get('chanceofrain', 0)
            daily_chance_of_rain = max(daily_chance_of_rain, int(chance_of_rain))

        chance_of_rain_list.append(daily_chance_of_rain)
        print(f"Date: {date}")
        print(f"Chance of Rain: {daily_chance_of_rain}%")
        print('---')

    # Plot the chance of rain for each day
    plt.figure(figsize=(10, 5))
    plt.plot(dates, chance_of_rain_list, marker='o')
    plt.title('Chance of Rain in Cairo for the Next 7 Days')
    plt.xlabel('Date')
    plt.ylabel('Chance of Rain (%)')
    plt.grid(True)
    plt.show()
```
